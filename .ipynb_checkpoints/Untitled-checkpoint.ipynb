```python
from fastbook import *
from fastai.vision.all import *
import pandas as pd
```

```python
path = '/home/cp47/CS344FinalProject/data/'
```

```python
emotions = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}
```

```python
df = pd.read_csv(path + 'icml_face_data.csv')
df.info()
```

```python
df[' Usage'].unique()
```

```python
def get_x(data): 
    images = np.fromstring(data[' pixels'], dtype=int, sep=' ')
    images = np.reshape(images, (48, 48)).astype(np.uint8)
    return images
def get_y(data): return emotions[data['emotion']]
```

```python
def splitter(data):
    train = data.index[data[' Usage'] == 'Training'].tolist()
    valid = data.index[data[' Usage'] != 'Training'].tolist()
    return train, valid
```

```python
dblock = DataBlock(blocks=(ImageBlock(cls=PILImage), CategoryBlock),
                   get_x = get_x,
                   get_y = get_y,
                   splitter=splitter)
```

```python
dsets = dblock.datasets(df)
```

```python
dls = dblock.dataloaders(df)
```

```python
dls.train.show_batch()
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(2, 6.3e-05)
```

```python
interp = Interpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9)
```

```python
def prepare_data(data):
    """ Prepare data for modeling 
        input: data frame with labels und pixel data
        output: image and label array """
    
    image_array = np.zeros(shape=(len(data), 48, 48))
    image_label = np.array(list(map(int, data['emotion'])))
    
    for i, row in enumerate(data.index):
        image = np.fromstring(data.loc[row, ' pixels'], dtype=int, sep=' ')
        image = np.reshape(image, (48, 48))
        image_array[i] = image
        
    return image_array, image_label
```

```python
train_image_array1, train_image_label1 = prepare_data(df[df[' Usage']=='Training'])
val_image_array, val_image_label = prepare_data(df[df[' Usage']=='PrivateTest'])
test_image_array, test_image_label = prepare_data(df[df[' Usage']=='PublicTest'])
```
